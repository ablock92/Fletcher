#Load articles and organize into monthlong segments

```python
client = MongoClient('mongodb://localhost:27017/')
db = client.Fletcher_db
collection = db.presidential_articles
```

```python
articles = []
cursor = collection.find()
for c in cursor:
    articles.append(c)
```

```python
year_months = defaultdict(list)
for a in articles:
    year_months[(a['Year'],a['Month'])].append(a)
```

##Concatenate the headlines for each month into one string

```python
stop_phrases = ['news summary','news analysis','business digest','early word','inside the times']

ym_headlines = defaultdict(str)
for key in year_months.keys():
    for article in year_months[key]:
        include = True
        headline = article['headline']['main'].lower()
        for phrase in stop_phrases:
            if phrase in headline:
                include = False
                break
        if include:
            ym_headlines[key] += headline + ' '
            
```

##Define the stopwords to remove from consideration

```python
stop = stopwords.words('english')
pres_names = ['obama','bush','clinton','bush','reagan']
dates = [str(year) for year in range(1980,2015)]

stop_all = stop + pres_names + dates
```

##DBScan Clusters


```python
vectorizer = TfidfVectorizer(stop_words=stop_all, ngram_range=(1,1),max_df=0.8,max_features=5000)
headlines_vectors = vectorizer.fit_transform(ym_headlines.values())
```

```python
dbscan = DBSCAN(eps=0.6,min_samples=3,metric='cosine',algorithm='brute')
clusters = dbscan.fit_predict(headlines_vectors.todense())
```

```python
cluster_dict = defaultdict(list)
for i,h in enumerate(ym_headlines.keys()):
    cluster_dict[clusters[i]].append(h)
    
combined_dict = defaultdict(str)
for cluster in cluster_dict.keys():
    for ym in cluster_dict[cluster]:
        combined_dict[cluster] += ym_headlines[ym] + ' '
```

```python
c_vectorizer = TfidfVectorizer(stop_words=stop_all, ngram_range=(1,1), max_features=5000,max_df=0.8)
cluster_vectors = c_vectorizer.fit_transform(combined_dict.values())
```

```python
cluster_kws = defaultdict(list)
for c in cluster_dict.keys():
    indices = cluster_vectors[c].indices
    data = cluster_vectors[c].data
    for kw_index,tfidf in zip(indices,data):
        #kw_index = cluster_vectors[c].indices[i]
        cluster_kws[c].append((tfidf,c_vectorizer.get_feature_names()[kw_index]))
    cluster_kws[c]=sorted(cluster_kws[c],reverse=True)
```
