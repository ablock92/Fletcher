```python
def format_date(date):
    date = date.split('T')[0]
    return date.replace('-','')
```

```python
pages_allowed = 100
articles_per_page = 10
initial_articles = api.search(fq = 'subsection_name:("Politics")')
hits = initial_articles['response']['meta']['hits']
```

```python
end_date = '20141016'
num = 0
while hits!=0:
    if hits < ((pages_allowed+1)*articles_per_page):
        pages_allowed = (hits)/articles_per_page
    for page in range(pages_allowed+1):
        articles = api.search(fq = 'subsection_name:("Politics")',page=page,end_date=end_date)
        articles_insert = articles['response']['docs']
        try:
            collection.insert(articles_insert,continue_on_error=True)
        except:
            pass
        time.sleep(0.1)
        
    hits = articles['response']['meta']['hits']
    print 'hits:',hits
    if hits>((pages_allowed+1)*articles_per_page):
        hits=0
        end_date = format_date(articles['response']['docs'][-1]['pub_date'])
        print 'end_date:',end_date
    print
```
