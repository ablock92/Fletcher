```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client.Fletcher_db
collection = db.presidential_articles
```

```python
articles = []
cursor = collection.find()
for c in cursor:
    articles.append(c)
```
