```python
query = "playoffs"
max_tweets = 500
search_results = tweepy.Cursor(api.search,q=query).items(max_tweets)
```

```python
tweets = []
for i,tweet in enumerate(search_results):
    tweets.append(tweet)
    print i, tweet.text
    #print i,tweet.retweeted
```

```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client.playoff_db
collection = db.playoffs

for t in tweets:
    p_dict={}
    p_dict['tweet'] = t.text
    p_dict['user'] =  t.user.name
    p_dict['favorite_count'] = t.favorite_count
    p_dict['retweet_count'] = t.retweet_count
    collection.insert(p_dict)
```
