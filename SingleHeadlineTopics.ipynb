```python
%matplotlib inline
import matplotlib.pyplot as plt
from pymongo import MongoClient

from collections import Counter
from collections import defaultdict

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import MiniBatchKMeans,KMeans,DBSCAN,AgglomerativeClustering
from sklearn.feature_selection import SelectKBest

from nltk.corpus import stopwords
from nltk import word_tokenize

from gensim import corpora, models, similarities
```

##Load the data from MongoDB

```python
client = MongoClient('mongodb://localhost:27017/')
db = client.Fletcher_db
collection = db.presidential_articles
```

```python
articles = []
cursor = collection.find()
for c in cursor:
    articles.append(c)
```

##Create a list of headlines
Filter out headlines of a certain time-specific format

```python
stop_phrases = ['news summary','news analysis','business digest','early word','inside the times','breakfast menu',
                'quotations day','quotations of the day','best sellers','evening briefing','monday briefing','tuesday briefing',
                'wednesday briefing','thursday briefing','friday briefing','saturday briefing','sunday briefing',
                'morning skim']

headlines=[]
ym_list = []
for article in articles:
    headline = article['headline']['main'].lower()
    include = True
    for phrase in stop_phrases:
        if phrase in headline:
            include = False
            break
    if include:
        headlines.append(headline)
        ym_list.append((article['Year'],article['Month']))
```

##Create topics from the headlines

```python
stop = stopwords.words('english')
pres_names = ['obama','bush','clinton','bush','reagan']
dates = [str(year) for year in range(1980,2015)]

stop_all = stop + pres_names + dates
```

```python
tokens = [word_tokenize(headline) for headline in headlines]
tokens = [[word for word in headline if word not in stop_all] for headline in tokens]

dictionary = corpora.Dictionary(tokens)
corpus = [dictionary.doc2bow(headline) for headline in tokens]

tfidf = models.TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]
```

```python
lda = models.ldamodel.LdaModel(corpus_tfidf, id2word=dictionary, num_topics=100,passes=30)
corpus_lda = lda[corpus_tfidf]
```

##Split articles into year/month segments
Concatenate the headlines into one string

```python
year_months = defaultdict(list)
for a in articles:
    year_months[(a['Year'],a['Month'])].append(a)
```

```python
ym_headlines = defaultdict(str)
for key in year_months.keys():
    for article in year_months[key]:
        ym_headlines[key]+=article['headline']['main']+' '
```

```python
ym_values = []
ym_keys = ym_headlines.keys()
for key in ym_keys:
    ym_values.append(ym_headlines[key])

ym_tokens = [word_tokenize(headlines_string) for headlines_string in ym_values]
ym_tokens = [[word for word in headlines_tokens if word not in stop_all] for headlines_tokens in ym_tokens]
```

##Check topic probabilities on each year/month segments

```python
ym_corpus = [dictionary.doc2bow(doc) for doc in ym_tokens]
ym_tfidf = models.TfidfModel(ym_corpus)
ym_corpus_tfidf = ym_tfidf[ym_corpus]
ym_corpus_lda = lda[ym_corpus_tfidf]
```

```python
ym_lda_probs=list(ym_corpus_lda)
```

##Map each topic to year/month and topic probability

```python
topics = defaultdict(list)
for ym,data in zip(ym_keys,ym_lda_probs):
    for d in data:
        topics[d[0]].append((ym,d[1]))

for topic in topics.keys():
    topic_yms = [value[0] for value in topics[topic]]
    for ym in ym_keys:
        if ym not in topic_yms:
            topics[topic].append((ym,0))
            
for t in topics.keys():
    topics[t]=sorted(topics[t])
```

```python
# y_values = [value[1] for value in topics[0]]
# plt.plot(range(len(topics[0])),y_values)
fig,ax=plt.subplots(figsize=(15,7))
y_values = [value[1] for value in topics[4]]

average_vals=[y_values[0]]
for i in range(1,len(y_values)-1):
    previous_val = y_values[i-1]
    current_val = y_values[i]
    next_val = y_values[i+1]
    average = (previous_val+current_val+next_val)/3.0
    average_vals.append(average)
average_vals.append(y_values[-1])

#ax.plot(range(len(topics[0])),y_values,color='r')
ax.plot(range(len(topics[0])),average_vals,color='r')


ticks = [b[0] for i,b in enumerate(sorted(ym_keys)) if i%12==0]
locs = [b for b in range(len(y_values)) if b%12==0]
_=ax.set_xticks(locs)
_=ax.set_xticklabels(ticks,rotation='vertical')
```

```python
import operator
lda_probs=list(corpus_lda)
```

```python
most_common = defaultdict(str)
ym_clusters = defaultdict(list)
i=0
for h,prob in zip(headlines,lda_probs):
    if len(prob)!=0:
        top = sorted(prob,key=operator.itemgetter(1),reverse=True)[0][0]
        most_common[top] += h + ' '
        ym_clusters[ym_list[i]].append(top)
    i+=1
```

##Find the top kws for each topic

```python
vectorizer = TfidfVectorizer(stop_words=stop_all, ngram_range=(1,2),max_df=0.8,max_features=5000)
headlines_vectors = vectorizer.fit_transform(most_common.values())
```

```python
kws = defaultdict(list)
for t in most_common.keys():
    indices = headlines_vectors[t].indices
    data = headlines_vectors[t].data
    for kw_index,tfidf in zip(indices,data):
        kws[t].append((tfidf,vectorizer.get_feature_names()[kw_index]))
    kws[t]=sorted(kws[t],reverse=True)
```

```python
##For 1 month average (Otherwise same functionality as below cell)


ym_headline_probas = defaultdict(dict)

for ym in ym_clusters.keys():
    curr_headline_probas = defaultdict(float)
    
    topic_list = ym_clusters[ym]
    num_headlines = len(topic_list)
    
    for t in topic_list:
        curr_headline_probas[t]+=1
    
    for t in curr_headline_probas.keys():
        curr_headline_probas[t]=float(curr_headline_probas[t])/(num_headlines)
    
    ym_headline_probas[ym] = curr_headline_probas
```
