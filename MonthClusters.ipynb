```python
%matplotlib inline
import matplotlib.pyplot as plt
from pymongo import MongoClient

from collections import Counter
from collections import defaultdict

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import MiniBatchKMeans,KMeans,DBSCAN,AgglomerativeClustering
from sklearn.feature_selection import SelectKBest

from nltk.corpus import stopwords
```

#Load articles and organize into monthlong segments

```python
client = MongoClient('mongodb://localhost:27017/')
db = client.Fletcher_db
collection = db.presidential_articles
```

```python
articles = []
cursor = collection.find()
for c in cursor:
    articles.append(c)
```

```python
year_months = defaultdict(list)
for a in articles:
    year_months[(a['Year'],a['Month'])].append(a)
```

##Concatenate the headlines for each month into one string

```python
ym_headlines = defaultdict(str)
for key in year_months.keys():
    for article in year_months[key]:
        ym_headlines[key]+=article['headline']['main']+' '
```

```python
vectorizer = TfidfVectorizer(stop_words="english", ngram_range=(1,1))
headlines_vectors = vectorizer.fit_transform(ym_headlines.values())
```

##Plot the inertia as the number of clusters ranges from 1 to 20

```python
inertia = []
for n_clusters in range(1,20):
    k_means = MiniBatchKMeans(n_clusters=n_clusters)
    clusters = k_means.fit_predict(headlines_vectors)
    inertia.append(k_means.inertia_)

```

```python
plt.plot(range(1,20),inertia)
```

##Build a Kmeans clusterer on the monthly headlines
Number of clusters (5) suggested by the 'elbow' in the inertia plot

```python
k_means = MiniBatchKMeans(n_clusters=5)
clusters = k_means.fit_predict(headlines_vectors)
```

```python
cluster_dict = defaultdict(list)
for i,h in enumerate(ym_headlines.keys()):
    cluster_dict[clusters[i]].append(h)
```

```python
combined_dict = defaultdict(str)
for cluster in cluster_dict.keys():
    for ym in cluster_dict[cluster]:
        combined_dict[cluster] += ym_headlines[ym] + ' '
```

###Name the Clusters from their highest scoring tfidf keywords

```python
c_vectorizer = TfidfVectorizer(stop_words="english", ngram_range=(1,1), max_features=1000)
cluster_vectors = c_vectorizer.fit_transform(combined_dict.values())
```

```python
cluster_kws = defaultdict(list)
for c in cluster_dict.keys():
    indices = cluster_vectors[c].indices
    data = cluster_vectors[c].data
    for kw_index,tfidf in zip(indices,data):
        #kw_index = cluster_vectors[c].indices[i]
        cluster_kws[c].append((tfidf,c_vectorizer.get_feature_names()[kw_index]))
    cluster_kws[c]=sorted(cluster_kws[c],reverse=True)
```

```python
for c in cluster_kws.keys():
    print c, [kw[1] for kw in cluster_kws[c][:5]]
```

##Remove the names of the Presidents and Vice-Presidents and recluster the articles

```python
stop = stopwords.words('english')
names_to_remove = ['obama','biden','bush','cheney','clinton','gore','bush','quayle','reagan']
stop += names_to_remove
```

```python
vectorizer = TfidfVectorizer(stop_words=stop, ngram_range=(1,1))
headlines_vectors = vectorizer.fit_transform(ym_headlines.values())
```

```python
inertia = []
for n_clusters in range(1,30):
    k_means = MiniBatchKMeans(n_clusters=n_clusters,n_init=20)
    clusters = k_means.fit_predict(headlines_vectors)
    inertia.append(k_means.inertia_)
```

```python
plt.plot(range(1,30),inertia)
plt.plot([0,30],[340,260])
```
